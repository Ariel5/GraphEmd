```python
import numpy as np
import pandas as pd
import random
import time
```

```python
n=2000
K=3
#Create empty matrix A
A = np.zeros((n,n))
```

```python
#Create y
np.random.seed(1)
y = np.random.choice([1,2,3], size=(n,1), p=[0.2,0.3,0.5])
```

```python
#Create X
np.random.seed(2)
X = np.random.uniform(0.0, 1.0, (n,1))
```

```python
#Change value of X
np.random.seed(3)
for i in range(0,n):
    if y[i] == 1:
        X[i,:] = np.random.beta(1,3)
    elif y[i] == 2:
        X[i,:] = np.random.beta(3,3)
    else:
        X[i,:] = np.random.beta(3,1)
```

```python
#Create P
P=np.matmul(X,X.transpose())
```

```python
P = P/np.amax(np.amax(P))
```

```python
np.random.seed(5)
for i in range(0,n):
    for j in range(1,n):
        A[i,j]=np.random.binomial(2,P[i,j])
```

```python
for i in range(0,n):
    for j in range(0,i):
        A[i,j] = A[j,i]
        A[i,i]=0
A
```

# Split Data

```python
from sklearn.model_selection import RepeatedKFold
rkf = RepeatedKFold(n_splits=5, n_repeats=10, random_state=5)
for train_index, test_index in rkf.split(A):
     print("TRAIN:", train_index, "TEST:", test_index)
     X_train, X_test = A[train_index], A[test_index]
     y_train, y_test = y[train_index], y[test_index]
     X1 = X_train.transpose()
     X_train, X_test = X1[train_index], X1[test_index]
```

```python
#Create nk
nk = np.zeros((1,K))
for i in range(0,K):
    nk[0,i]=((i+1)*(i+2))/2
```

```python
#Create w
w = np.zeros((int(n/5*4),K))
for i in range(0,int(n/5*4)):
    k=int(y_train[i])
    w[i][k-1]=1/nk[0,k-1]
```
